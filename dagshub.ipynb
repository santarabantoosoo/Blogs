Let's imagine that we have on the left a huge file. It's available locally and we need to track it. We have both git and dvc to track the file. 

[![Screenshot-from-2022-08-26-15-53-05.png](https://i.postimg.cc/rF6zdpV9/Screenshot-from-2022-08-26-15-53-05.png)](https://postimg.cc/23wCMrBq)

We use `dvc add` to track the data using dvc. This also does two other modifications. It adds the data to gitignore so that it is not tracked by git. In addition, it creates a .dvc file that works as a pointer to the data version and is tracked by git. This file will contain metadata about the version of the data. 

[![Screenshot-from-2022-08-26-16-04-09.png](https://i.postimg.cc/fR9GVwbY/Screenshot-from-2022-08-26-16-04-09.png)](https://postimg.cc/fVM2FQmb)

Things got complicated now. We have multiple versions of our datasets that are tracked by dvc and git. We compared the performance metrics of different experiments that were based on different versions of a dataset. We now know which version is offering the best performance. However, how can we reach them again. 

First: decide how these versions were tracked by dvc (locally created and then pushed to dvc or created remotely on another computer and pushed to dvc) 

Second: 
- If created locally; use `dvc checkout`   
- If created on another remote machine, use `dvc pull`
