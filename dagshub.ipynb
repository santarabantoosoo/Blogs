* you have continuous integration where you automatically test your models 

So DVC is important.

Dagshub comes to the rescue! It solves this with 4 lines of code and it offers 10 GB of storage

Let's imagine that we have on the left a huge file. It's available locally and we need to track it. We have both git and dvc to track the file. 

[![Screenshot-from-2022-08-26-15-53-05.png](https://i.postimg.cc/rF6zdpV9/Screenshot-from-2022-08-26-15-53-05.png)](https://postimg.cc/23wCMrBq)

We use `dvc add` to track the data using dvc. This also does two other modifications. It adds the data to gitignore so that it is not tracked by git. In addition, it creates a .dvc file that works as a pointer to the data version and is tracked by git. This file will contain metadata about the version of the data. 

[![Screenshot-from-2022-08-26-16-04-09.png](https://i.postimg.cc/fR9GVwbY/Screenshot-from-2022-08-26-16-04-09.png)](https://postimg.cc/fVM2FQmb)

Things got complicated now. We have multiple versions of our datasets that are tracked by dvc and git. We compared the performance metrics of different experiments that were based on different versions of a dataset. We now know which version is offering the best performance. However, how can we reach them again. 

First: decide how these versions were tracked by dvc (locally created and then pushed to dvc or created remotely on another computer and pushed to dvc) 

Second: 
- If created locally; use `dvc checkout`   
- If created on another remote machine, use `dvc pull`


## DagsHub getting started 
> I am mainly following this [playlist](https://www.youtube.com/watch?v=ECbVxGqS0f0&list=PLC4-OBA5bH3hwxzqytYoO8I0T-Yj_SglD) which is based on this [getting started guide](https://dagshub.com/docs/getting_started/overview/) ==> pretty straight foreward 


**Side note** 

In part 3 of the getting started guide, you will see experiment logging using DagsHub. You will see an amazing experiment tab where you can keep track of models, params and metrics. 
However, they never mention MLFLOW. Is this a seamless integration? Are we using MLFLOW without even installing it? 



[![Screenshot-from-2022-08-26-17-09-30.png](https://i.postimg.cc/J04dr8yV/Screenshot-from-2022-08-26-17-09-30.png)](https://postimg.cc/cg2chVMX)

No. not really. DagsHub has a built-in experiment tracking. However, it has a major disadvantage. It doesn't allow autolog for all ML libraries. Automatic logging allows you to log metrics, parameters, and models without the need for explicit log statements.

[Complete Guide to Experiment Tracking With MLflow and DagsHub | by Bex T. | Towards Data Science](https://towardsdatascience.com/complete-guide-to-experiment-tracking-with-mlflow-and-dagshub-a0439479e0b9)

For example, MLFLOW supports all of these autologs

Scikit-learn | TensorFlow and Keras | Gluon | XGBoost | LightGBM | Statsmodels | Spark | Fastai | Pytorch

[MLFLOW_autologging](https://mlflow.org/docs/latest/tracking.html#id2)

On the other hand, native DagsHub can support autologging only for these: 

- pytorch lightening
- fastai
- keras 

[DagsHub Tracking - DagsHub Docs](https://dagshub.com/docs/feature_guide/git_tracking/)
